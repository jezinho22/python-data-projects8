```python
import pandas as pd

#fetch file of results scraped from webpages
dfresults = pd.read_csv ('eplscores plus halftimes.csv').fillna('0')

#function to work out points for results
def resultpoints (hresult, aresult):

    if hresult == aresult:
        return 1
    elif hresult > aresult:
        return 3
    elif hresult < aresult:
        return 0

#get columns of points for each match by applying function
dfresults ['Homepoints'] = dfresults.apply(lambda x: resultpoints(x ['homeresult'],x ['awayresult']),axis=1)
dfresults ['Awaypoints'] = dfresults.apply(lambda x: resultpoints(x ['awayresult'],x ['homeresult']),axis=1)

dfresults ['HTHomepoints'] = dfresults.apply(lambda x: resultpoints(x ['homehalftime'],x ['awayhalftime']),axis=1)
dfresults ['HTAwaypoints'] = dfresults.apply(lambda x: resultpoints(x ['awayhalftime'],x ['homehalftime']),axis=1)

#aggregate matches into total points for each team
hpoints = dfresults.groupby('home').Homepoints.sum().reset_index(name="Home_FT_pts").rename(columns={'home':'team'})
apoints = dfresults.groupby('away').Awaypoints.sum().reset_index(name="Away_FT_pts").rename(columns={'away':'team'})

HThpoints = dfresults.groupby('home').HTHomepoints.sum().reset_index(name="Home_HT_pts").rename(columns={'home':'team'})
HTapoints = dfresults.groupby('away').HTAwaypoints.sum().reset_index(name="Away_HT_pts").rename(columns={'away':'team'})

#create a dataframe with the aggregated data
ap3 = hpoints.merge(apoints, on ='team').merge(HThpoints, on ='team' ).merge(HTapoints, on ='team')

#add columns to add up the points
ap3['FT_Points'] = ap3.Home_FT_pts + ap3.Away_FT_pts
ap3['HT_Points'] = ap3.Home_HT_pts + ap3.Away_HT_pts
ap3['Diff'] = ap3.FT_Points - ap3.HT_Points

#print table sorted by half time points
ap3.sort_values('HT_Points', ascending = False)
```

So Norwich would be comfortably mid table, and within reach of European places, if all the matches had stopped at half time.

```python
#print table sorted by difference between half time and full time points
ap3.sort_values('Diff', ascending = False)
```
