```python
import pickle
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import time
import random
import pandas as pd
import re

with open('korf2appearances.txt', 'rb') as handle:
    dictappearances = pickle.loads(handle.read())
with open('korf2scorers.txt', 'rb') as handle:
    dictgoals = pickle.loads(handle.read())
with open ('korf2.txt', 'rb') as handle:
    dictmatches = pickle.loads(handle.read())
    
lstteams = dictmatches['teams']


#print (dictgoals['Tornadoes 2'])
#display (dictgoals['Tornadoes 2'])
#display (dictappearances['Norwich Knights 2'])
```

```python
dictplayerstats = {}

for team in lstteams:
    dfappearances = dictappearances[team]
    dftopscorers = dictgoals[team]
    dfplayerstats = pd.merge (dfappearances, dftopscorers, how = 'outer')
    #clean up NaNs
    dfplayerstats['goals'].fillna(0, inplace=True)
    #add avge column & sort
    dfplayerstats['avge'] = round(dfplayerstats.goals / dfplayerstats.appearances, 1)
    dfplayerstats.sort_values(by = ['avge'], ascending = False, inplace = True)
    #add to a dictonary
    dictplayerstats[team] = dfplayerstats
dictplayerstats['index'] = lstteams

with open('korf2playerstats.txt', 'wb') as handle:
    pickle.dump(dictplayerstats, handle)
```
