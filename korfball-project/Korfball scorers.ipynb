```python
import pickle
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import time
import random
import pandas as pd
import re


filename = 'Norfolk2_match_facts'

with open(f'{filename}.txt', 'rb') as handle:
    dictmatches = pickle.loads(handle.read())

dictteamscorers = {}

lstteams = dictmatches['teams']

for team in lstteams:
    dictteamscorers[team] = []
    
for match in dictmatches['index']:
    dfmatch2 = dictmatches[match]
    
    lstgoalseach = []

    #work in columns 4 and 5 - scorer columns - make two new dfs
    for x in [5,6]:
        #count up goals using groupby and add to list of each team's scorers
        team = dfmatch2.columns[x].replace("_scorers", "")
        dfgoalseach = dfmatch2.groupby(dfmatch2.columns[x]).index.count().reset_index().rename(columns = {'index': 'goals', f'{team}_scorers':team})
        dictteamscorers[team].append(dfgoalseach)            
        
with open('Norfolk2_goals.txt', 'wb') as handle:
  pickle.dump(dictteamscorers, handle)

```

```python
#combine scorers for each club into one df, group & sum to add up all goals scored
dicttopscorers = {}
for team in lstteams:
    dfnew = pd.concat(dictteamscorers[team])
    dfnewest = dfnew.groupby(dfnew.columns[0]).goals.sum().reset_index()
    dfnewest = dfnewest.sort_values(by=['goals'],ascending = False)
    dicttopscorers [team] = dfnewest

with open('Norfolk2_scorers.txt', 'wb') as handle:
  pickle.dump(dicttopscorers, handle)
```
