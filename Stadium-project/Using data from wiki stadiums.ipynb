```python
#import csv and check
stadia = pd.read_csv('Eng stadia utf.csv')
stadia.to_csv('Eng stadia utf2.csv')

#leading spaces are messing up filtering
stadia = stadia.applymap(lambda x: x.strip() if type(x) is str else x)
stadia.rename({'League_Pos':'Rank_in_league'}, axis = 1, inplace = True)

stadia.info()
stadia.head(15)
```

```python
#get stadia with over 20000 capacity
biggest_stadia = stadia[stadia.Capacity > 20000]
biggest_stadia[['Stadium','Capacity']].head(10)
```

```python
stadia['League_Tier'].unique()
```

```python
champs = stadia[stadia.League_Tier == 'Championship']
champs[['Stadium', 'Capacity', 'Club', 'League_Pos']].sort_values('Capacity', ascending = False)
```
